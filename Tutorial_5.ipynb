```python
#Libraries Required
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
from rpeakdetectr.rpeakdetect import detect_beats, plot_peak_detection
import wfdb
```

```python
def RR(key,ecg,sampling_freq):
    RR_peak = plot_peak_detection(ecg,sampling_freq)
    plt.stem(RR_peak)
    plt.title(key)
    plt.show()
    
```

```python
def rpeak(key,ecg):
    sampling_freq= 300 #ITs given in A00008.hea file
    R_peak = (detect_beats(ecg[0],sampling_freq))
    plt.stem(R_peak)
    plt.title(key)
    # plt.scatter(R_peak[:-1], R_peak[1:])
    plt.show()
    
    
```

```python
def ECG(key,adr):
    mat = scipy.io.loadmat(adr)
    ecg = mat.get('val')
    plt.plot(ecg[0])
    plt.title(key)
    plt.show()
    rpeak(key,ecg)
```

```python
file_name = {
    "ecg_normal":'A00001.mat',
    "ecg_noisy":'A00022',
    "ecg_af":'A00004',
    "ecg_others":'A00008'
}
```

```python
for key,adr in file_name.items():
    ECG(key,adr)
```
